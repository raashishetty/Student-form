```python
import tkinter
from tkinter import *
import pymysql
top=tkinter.Tk()
top.geometry('800x450')
label=Label(top,text="Enter Student ID: ",font=('arial',30),fg='blue')
label.grid(row=0,column=0)
E=Entry(top,bd=5,width=50)
E.grid(row=0,column=1)

label1=Label(top,text="Enter Student Name: ",font=('arial',30),fg='blue')
label1.grid(row=1,column=0)
E1=Entry(top,bd=5,width=50)
E1.grid(row=1,column=1)

label2=Label(top,text="Enter Student Address: ",font=('arial',30),fg='blue')
label2.grid(row=2,column=0)
E2=Entry(top,bd=5,width=50)
E2.grid(row=2,column=1)

label3=Label(top,text="Date of joining: ",font=('arial',30),fg='blue')
label3.grid(row=3,column=0)
E3=Entry(top,bd=5,width=50)
E3.grid(row=3,column=1)

label4=Label(top,text="Field: ",font=('arial',30),fg='blue')
label4.grid(row=4,column=0)
E4=Entry(top,bd=5,width=50)
E4.grid(row=4,column=1)

def Button(selection):
    print("Student ID is: ",E.get())
    print("Student Name is: ",E1.get())
    print("Student Address is: ",E2.get())
    print("Date of joining: ",E3.get())
    print("Field: ",E4.get())
    
    ID=E.get()
    Name=E1.get()
    Address=E2.get()
    Join=E3.get()
    Field=E4.get()
    
    if selection in ('Save'):
        
        connection=pymysql.connect(host="localhost",user="root",
                                   passwd="P@ssw0rd",database="test")
        curs=connection.cursor()
        #curs.execute("select version()")
        #data=curs.fetchone()
        #print("Mysql Database version ",data)

        query1="create table if not exists student(id char(20) primary key not null,\
                            name char(20),address char(20) )"
        query2="create table if not exists stud_det(id char(20) primary key not null,\
                            dj DATE,field char(20) )"
        
        curs.execute(query1)
        curs.execute(query2)
        connection.commit()
        print("Table created succesfully")

        inser1="insert into student (id,name,address)\
        values('%s','%s','%s')"%(ID,Name,Address)
        inser2="insert into stud_det (id,dj,field)\
        values('%s','%s','%s')"%(ID,Join,Field)

        try:
            curs.execute(inser1) 
            curs.execute(inser2)
            connection.commit()
            print("Details saved ",ID,",",Name,",",Address,",",Join,",",Field)
            connection.close()
        except Error as e:
            print("Error occured", e)
            connection.rollback()
            connection.close()

    elif selection in ('Update'):
        query1="update student set name='%s'"%(Name)+",\
                address='%s'"%(Address)+" where id='%s'"%(ID)
        query2="update stud_det set field='%s'"%(Field)+",\
                dj='%s'"%(Join)+" where id='%s'"%(ID)
        connection=pymysql.connect(host="localhost",user="root",
                               passwd="P@ssw0rd",database="test")
        curs=connection.cursor()
        curs.execute(query1)
        curs.execute(query2)
        connection.commit()
        connection.close()
        print("Updated successfully ", ID)
    elif selection in ('Delete'):
        query1="delete from student where id='%s'"%(ID)+"or name='%s'"%(Name)
        query2="delete from stud_det where id='%s'"%(ID)
        connection=pymysql.connect(host="localhost",user="root",
                               passwd="P@ssw0rd",database="test")
        curs=connection.cursor()
        curs.execute(query1)
        curs.execute(query2)
        connection.commit()
        connection.close()
        print("Deleted successfully ", ID)
    elif selection in ('Select'):
        query="select student.id,student.name,student.address,stud_det.dj,stud_det.field \
                from student inner join stud_det where student.id='%s'"%(ID)
        connection=pymysql.connect(host="localhost",user="root",
                               passwd="P@ssw0rd",database="test")
        curs=connection.cursor()
        curs.execute(query)
        row=curs.fetchall()
        address1=''
        name1=''
        ID1=''
        join1=''
        field1=''
        for r in row:
            ID1=r[0]
            name1=r[1]
            address1=r[2]
            join1=r[3]
            field1=r[4]
        E.delete(0,END)
        E1.delete(0,END)
        E2.delete(0,END)
        E3.delete(0,END)
        E4.delete(0,END)
        E.insert(0,ID1)
        E1.insert(0,name1)
        E2.insert(0,address1)
        E3.insert(0,join1)
        E4.insert(0,field1)
       
        connection.close()
        print("Selected successfully ", ID)
        

but_ins=tkinter.Button(text="Save",fg='white',bg='black',
                       font=('arial',20,'bold'),command=lambda:Button('Save'))
but_ins.grid(row=5,column=0)

but_up=tkinter.Button(text="Update",fg='white',bg='black',
                      font=('arial',20,'bold'),command=lambda:Button('Update'))
but_up.grid(row=5,column=1)

but_del=tkinter.Button(text="Delete",fg='white',bg='black',
                       font=('arial',20,'bold'),command=lambda:Button('Delete'))
but_del.grid(row=7,column=0)

but_sel=tkinter.Button(text="Select",fg='white',bg='black',
                       font=('arial',20,'bold'),command=lambda:Button('Select'))
but_sel.grid(row=7,column=1)

top.mainloop()
```
